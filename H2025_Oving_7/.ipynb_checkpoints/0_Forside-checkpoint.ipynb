#### Newton's Metode Illustrert
Newton's metode er er en metode som raskt kan regne seg frem til løsningene på ligningen $f(x) = 0$ gitt at den deriverte $f'(x)$ er kjent. Nedenfor ser du et interaktivt eksempel der Newtons Metode brukes for å finne én av løsningene på ligningen $x^3 - 3x^2 + 4 = 0$:

```python
from numerikk_demoer import NewtonsMethodDemo
from numpy import cos, sin
%matplotlib widget

def f(x):
    return x**3 - 3*x**2 - x + 4

def df_dx(x):
    return 3*x**2 - 6*x - 1

NewtonsMethodDemo(f, df_dx, x0=0);
```
